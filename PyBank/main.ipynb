```python
# Pybank

# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path("budget_data.csv")

# Initialize dictionary of records
records = []

# Initialize the metric variables
count = 0
total = 0
average = 0
minimum = 0
maximum = 0

# Open the csv file as an object
with open(csvpath, 'r') as csvfile:
    
    # Create a csv reader object
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Skip the header row
    header = next(csvreader)
    
    # Loop through the rows in the file and store the data in records
    for row in csvreader:
        records.append(row)
        
# Extract Profits/losses from records
profits_losses = [int(row[1]) for row in records]

# Calculate the count and total
count = len(profits_losses)
total = sum(profits_losses)



# Calculate the change in profits/losses over the entire period
changes = [profits_losses[i+1]-profits_losses[i] for i in range(len(profits_losses)-1)]

# Calculate the average of the changes in Profit/Losses over the entire period
average_change = round(sum(changes)/len(changes), 2)

# Get the minimum and maximum change in profits/losses and their indices
minimum_change = min(changes)
minimum_index = changes.index(minimum_change)
maximum_change = max(changes)
maximum_index = changes.index(maximum_change)

# Get the month and year associated with the minimum and maximum changes in profits/losses
minimum_month_year = records[minimum_index+1][0]
maximum_month_year = records[maximum_index+1][0]

# Print the financial analysis results
print("Financial Analysis")
print("------------------------")
print(f"Total Months: {count}")
print(f"Total: ${total}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Profits: {maximum_month_year} (${maximum_change})")
print(f"Greatest Decrease in Profits: {minimum_month_year} (${minimum_change})")

# Set output file name
output_path = 'output.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write daily_average to the output file, convert to string
    file.write("Financial Analysis \n")
    file.write("------------------------ \n")
    file.write(f"Total Months: {count}\n")
    file.write(f"Total: ${total}\n")
    file.write(f"Average Change: ${average_change}\n")
    file.write(f"Greatest Increase in Profits: {maximum_month_year} (${maximum_change})\n")
    file.write(f"Greatest Decrease in Profits: {minimum_month_year} (${minimum_change})\n")

```
